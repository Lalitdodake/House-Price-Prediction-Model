```python
import pandas as pd
df=pd.read_csv("Housing.csv")
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
catogries=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']
df=pd.get_dummies(df,columns=catogries,drop_first=True)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
area1=scale.fit_transform(df[['area']])
df['area']=area1
```

```python
df
```

```python
x=df.drop(['price'],axis=1)
```

```python
x
```

```python
y=df['price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=50)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
Lrmodel=LinearRegression()
Lrmodel.fit(X_train,y_train)
```

```python
y_train_predict=Lrmodel.predict(X_train)
```

```python
y_test_predict=Lrmodel.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
# for training data
r2_score(y_train,y_train_predict)
```

```python
#for testing data
r2_score(y_test,y_test_predict)
```
